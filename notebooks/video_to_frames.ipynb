```python
import cv2
import os
import re
```

```python
root = '../data/'

count = 0
for file in os.listdir(root + 'videos/'):
    if file.endswith('.mp4'):
        folder = re.match(r"^[^\d]*", file).group(0)
        video = cv2.VideoCapture(root + 'videos/' + file)
        success, frame = video.read()
        while success:
            frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
            file_path = root + 'images/' + f'{folder}/frame_{count}.jpg'
            print(file_path)
            cv2.imwrite(file_path, frame)
            success, frame = video.read()
            count += 1

video.release()
```
