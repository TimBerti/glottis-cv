```python
import os
import cv2
import re
import numpy as np
```

```python
data_dir = '../data/'
train_dir = data_dir + 'images/train/'
valid = data_dir + 'images/valid/'
test_dir = data_dir + 'images/test/'

train_size = 0.7
valid_size = 0.2

count = 0
for file in os.listdir(data_dir + 'videos/'):
    if file.endswith('.mp4'):
        category = re.match(r"^[^\d]*", file).group(0)
        video = cv2.VideoCapture(data_dir + 'videos/' + file)

        success, frame = video.read()
        while success:

            if np.random.rand() < train_size:
                file_path = train_dir + category + '/' + str(count) + '.jpg'
            elif np.random.rand() < valid_size / (1 - train_size):
                file_path = valid + category + '/' + str(count) + '.jpg'
            else:
                file_path = test_dir + category + '/' + str(count) + '.jpg'

            print(file_path)
            cv2.imwrite(file_path, frame)

            success, frame = video.read()
            count += 1

        video.release()
```
