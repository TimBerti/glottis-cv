```python
import seaborn as sns
import pandas as pd
```

```python
open = [0, 2, 8, 4, 14, 3, 6]
closed = [1, 2, 4, 8, 7, 5, 4]
equal = [3, 5, 6, 2, 13, 6, 3]

open = [x / sum(open) for x in open]
closed = [x / sum(closed) for x in closed]
equal = [x / sum(equal) for x in equal]

segments = ['A', 'B', 'C', 'D', 'E', 'F', 'G']

df = pd.DataFrame({'Open': open, 'Closed': closed, 'Equalizing': equal, 'Segment':segments})

df = df.melt('Segment', var_name='State',  value_name='Frequency')

sns.barplot(x="Segment", y="Frequency", hue='State', data=df);
```
