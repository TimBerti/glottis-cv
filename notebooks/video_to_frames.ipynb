```python
import os
import cv2
import re
```

```python
data_dir = '../data/'
train_dir = data_dir + 'images/train/'
valid_dir = data_dir + 'images/valid/'

skip_frames = 10

count = 0
for file in os.listdir(data_dir + 'videos/'):
    if file.endswith('.mp4'):
        category = re.match(r"^[^\d]*", file).group(0)
        video = cv2.VideoCapture(data_dir + 'videos/' + file)

        if file.endswith('6.mp4'):
            path = valid_dir + category
        else:
            path = train_dir + category

        success, frame = video.read()
        while success:
            if count % skip_frames == 0:
                file_path = path + '/' + str(count) + '.jpg'
                print(file_path)
                cv2.imwrite(file_path, frame)

            success, frame = video.read()
            count += 1

        video.release()
```
